# Write a function to calculate the product of two matrices. Generate two random matrices $x \in \mathbf{R}^{m \times n}$ and $y \in \mathbf{R}^{n \times m}$ contains intergers between $[-l, l]$ and find the product of of $x$ and $y$ (your function should handle different order matrices). Check your result with the [numpy.matmul](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) or [numpy.dot](https://numpy.org/doc/stable/reference/generated/numpy.dot.html).




```python
m = 3
n = 5
l = 9
x = []
y = []

def matrix_multiplication(
    x: np.array,
    y: np.array,
)->np.array:
    """
    Dot product between two vectors
    
    Inputs:
        - x: 1st matrix
        - y: 2nd matrix
       
    Output:
        - val: multiplication x and y
    
    """
    
    val = []
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    
    if x.shape[1] == y.shape[0]:
        val = np.zeros((x.shape[0], y.shape[1]))
        for i in range(x.shape[0]):
            for j in range(y.shape[1]):
                for k in range(x.shape[1]):
                    val[i, j] += x[i, k] * y[k, j]
    else:
        raise ValueError('Please check the dimention of the matrix!')
            
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return val

# np.random.seed(42)# for reproduce the random numbers
A = np.random.randint(-l, l, [m, n]).astype(float)
B = np.random.randint(-l, l, [n, m]).astype(float)

print('Two random matrices are:')
print('A: \n{}\nand\nB: \n{}' .format(A, B))
print('-'*70)

val = matrix_multiplication(A, B)
print('The multiplication of A and B using my function is: \n{}' .format(val))
print('-'*70)

val1 = np.matmul(A, B)
print('The multiplication of A and B using numpy.matmul is: \n{}' .format(val1))
print('-'*70)

val1 = np.dot(A, B)
print('The multiplication of A and B using numpy.dot is: \n{}' .format(val1))
print('-'*70)
```

# Implement row and column sweep out and check your script with random example

```python
def sweep_out_row(
    A: np.array,
    ith_row: int,
    pivot_element: Tuple[int, int]
)->np.array:
    """
    Sweep out a i-th row based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_row: row to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant row sweep out matrix
    """
    
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    
    A = elementary_ops_scale_col(A, pivot_element[1], 1./A[pivot_element[0], pivot_element[1]])
    for j in range(A.shape[1]):
        if(j != pivot_element[1]):
            A = elementary_ops_change_col(A, j, pivot_element[1], -A[ith_row, j])
        
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A
    
def sweep_out_row_mult(
    A: np.array,
    ith_row: int,
    pivot_element: Tuple[int, int]
)->np.array:
    """
    Sweep out a i-th row based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_row: row to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant row sweep out matrix
    """
    
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    
    E_ops = np.eye(A.shape[1])
    A, C_i = elementary_ops_scale_col_mult(A, pivot_element[1], 1./A[pivot_element[0], pivot_element[1]])
    E_ops = C_i.copy()
    for j in range(A.shape[1]):
        if(j != pivot_element[1]):
            A, C_ij_scale = elementary_ops_change_col_mult(A, j, pivot_element[1], -A[ith_row, j])
            E_ops = matrix_multiplication(E_ops, C_ij_scale)
        
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A, E_ops
    
    
def sweep_out_col(
    A: np.array,
    ith_col: int,
    pivot_element: Tuple[int, int]
)->np.array:
    """
    Sweep out a i-th column based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_col: column to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant column sweep out matrix
    """
    
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    
    A = elementary_ops_scale_row(A, pivot_element[0], 1./A[pivot_element[0], pivot_element[1]])
    for i in range(A.shape[0]):
        if(i != pivot_element[0]):
            A = elementary_ops_change_row(A, i, pivot_element[0], -A[i, ith_col])
        
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A

def sweep_out_col_mult(
    A: np.array,
    ith_col: int,
    pivot_element: Tuple[int, int]
)->Tuple[np.array, np.array]:
    """
    Sweep out a i-th column based on the given pivot element
    
    Inputs:
        - A: Given matrix
        - ith_col: column to sweep out
        - pivot_element pivot element
        
    Output:
        - A: Resultant column sweep out matrix
        - E_ops: Product of all the elementary operations
    """
    
    ############################################################################
    #                             Your code will be here                       #
    #--------------------------------------------------------------------------#
    
    E_ops = np.eye(A.shape[0])
    A, I_i = elementary_ops_scale_row_mult(A, pivot_element[0], 1./A[pivot_element[0], pivot_element[1]])
    E_ops = I_i.copy()
    for i in range(A.shape[0]):
        if(i != pivot_element[0]):
            A, I_ij_scale = elementary_ops_change_row_mult(A, i, pivot_element[0], -A[i, ith_col])
            E_ops = matrix_multiplication(I_ij_scale, E_ops)
    #--------------------------------------------------------------------------#
    #                             End of your code                             #
    ############################################################################
    
    return A, E_ops




np.random.seed(42)# for reproduce the random numbers
m = 3
n = 5
A = np.random.randint(-l, l, [m, n]).astype(float)

print('A random matrices A:\n{}' .format(A))
print('-'*70)
i = 1
j = 2
scalar = 3
E_ij = sweep_out_row(np.copy(A), i, (i, j))
print('{}-th row sweep out based on ({}, {}) as pivot on A: \n{}\nis : \n{}' .format(i, i, j, A, E_ij))
print('-'*70)

E_ij, E_ops = sweep_out_row_mult(np.copy(A), i, (i, j))
print('{}-th row sweep out (multiplication) based on ({}, {}) as pivot on A: \n{}\nis: \n{}\nand product of all the elementary operations E_ops: \n{}' .format(i, i, j, A, E_ij, E_ops))
print('-'*70)

E_ij = sweep_out_col(np.copy(A), j, (i, j))
print('{}-th col sweep out based on ({}, {}) as pivot on A: \n{}' .format(j, i, j, E_ij))
print('-'*70)

E_ij, E_ops = sweep_out_col_mult(np.copy(A), j, (i, j))
print('{}-th col sweep out (multiplication) based on ({}, {}) as pivot on A: \n{}\n is : \n{}\nand product of all the elementary operations E_ops: \n{}' .format(j, i, j, A, E_ij, E_ops))
print('-'*70)

```

# Implement reduced echelon form of a matrix $A$
