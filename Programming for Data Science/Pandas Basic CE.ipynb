```python
s1 = titanic.groupby(['sex'])['fare'].count(); s1.index
```

```python
s2 = titanic.groupby(['sex'])['fare'].mean(); s2
```

```python
plt.figure(figsize=(5,5))
plots = plt.bar(s1.index,s1.values,color='pink')
plt.ylim(0,700)
for p,j,k in zip(plots,s1.index,s2.values):
    plt.annotate('mean fare = \n'+str(k),
                 xy=(p.get_x()+ p.get_width()/8, 
                     p.get_y()+ p.get_height()+10),
                 color='red',
                fontname="Times New Roman", size=15,
                 fontweight="bold",style='italic')
    #print(p.get_x())
plt.xlabel('sex',color='red',fontweight='bold')
plt.ylabel('frequency',color='red',fontweight='bold')
plt.title('Plot of frequency by gender',color='blue',fontweight='bold')
plt.xticks(color='green',fontweight='bold')
plt.yticks(color='green',fontweight='bold')
```

```python
pt1 = titanic.pivot_table('age',['sex','class'])
```

```python
pt_plot = pt1.plot(kind='bar')
pt_plot.legend(title='gender-class',bbox_to_anchor=(1, 1.02),
         loc='upper left')
```

```python
colors = ['C'+str(i) for i in range(len(pt1.index))]; colors
```

```python
from matplotlib.patches import Patch
```

```python
cmap = dict(zip(pt1.index, colors)); cmap
patches = [Patch(color=v, label=k) for k, v in cmap.items()]
```

```python
pt_plot = pt1.plot(kind='bar',color=colors)
#pt_plot.legend(title='gender-class',bbox_to_anchor=(1, 1.02),
#         loc='upper left')
pt_plot.legend(
        handles=patches, 
        bbox_to_anchor=(1.04, 0.5), 
        loc='center left', 
        borderaxespad=0,
        title='gender-class'
    )
bars = pt_plot.patches
for i,c in zip(bars,colors):
    i.set_facecolor(c)
```
